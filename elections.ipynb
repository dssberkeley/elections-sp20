### Getting the data from 2020: 

```python
# use pandas.read_html instead
polls20 = pd.read_html('https://www.realclearpolitics.com/epolls/latest_polls/democratic_nomination_polls/')

for t in polls20:
    display(t)
```

```python
poll20 = pd.DataFrame(columns=[0,1,2,3])

for t in polls20:
    if (len(t) != 1): 
        poll20 = poll20.append(t.iloc[1:])

poll20 = (poll20.rename(columns={0:'Race', 1:'Poll', 2:'Results', 3:'Spread'})
            .query("Race != '2020 Democratic Presidential Nomination'")
            .reset_index().iloc[:, 1:])
display(poll20)
```

### Creating new columns

```python
poll20['Spread'].str.extract(r'([A-z]+)')
```

```python
poll20['Advantage'] = poll20['Spread'].str.extract(r'([A-z]+)')
poll20.head()
```

```python
poll20['Advantage'].value_counts() / len(poll20)
```

Hmm, Sanders has a huge advantage here. However, it doesn't seem to hold for the Super Tuesday result... what may have gone wrong? 

### Grouping by States

```python
poll20['Race']
```

```python
poll20['Race'].str.extract(r'([\w ]+) Democratic')
```

```python
poll20['State'] = poll20['Race'].str.extract(r'([\w ]+) Democratic')
poll20.head()
```

```python
poll20['Advantage'].value_counts().index[0]
```

```python
def mode(series):
    return series.value_counts().index[0]
poll20_grouped = poll20[['State', 'Advantage']].groupby('State').agg(mode)
poll20_grouped
```

```python
poll20_grouped['Advantage'].value_counts() / len(poll20_grouped)
```
